**advance matplotlib**

```python
batters=pd.read_csv('/content/batter.csv')
batters.head()
```


**Subplot**

```python
# we can plot 2 graphs in single graph
fig,ax=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
ax[0].scatter(batters['avg'],batters['strike_rate'],color='green')
ax[1].scatter(batters['avg'],batters['runs'],color='blue')

ax[0].set_title('avg vs strike rate',color='red')
ax[1].set_title('avg vs runs',color='red')

# ax[0].set_xlabel('avg')
ax[0].set_ylabel('strike rate')

ax[1].set_xlabel('avg')
ax[1].set_ylabel('runs')
```

```python
# We can create 4 graphs in a single graph by using this methods
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='green')
ax[0,1].scatter(batters['avg'],batters['runs'],color='red')
ax[1,0].hist(batters['avg'],color='pink')
ax[1,1].hist(batters['runs'],color='yellow')

ax[0,0].set_title('avg vs strike rate',color='red')
ax[0,1].set_title('avg vs runs',color='red')
ax[1,0].set_title('avg ',color='red')
ax[1,1].set_title('runs',color='red')

ax[0,0].set_xlabel('avg')
ax[0,0].set_ylabel('strike rate')

ax[0,1].set_xlabel('avg')
ax[0,1].set_ylabel('runs')

ax[1,0].set_ylabel('strike rate')
ax[1,1].set_ylabel('runs')

fig.show()
```
